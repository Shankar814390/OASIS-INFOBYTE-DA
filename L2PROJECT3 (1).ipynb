## **HANDLING IMBALANCED DATASET**

```python
normal = df[df['Class']==0]
fraud = df[df['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=473)
normal_sample.shape
```

```python
new_data = pd.concat([normal_sample,fraud],ignore_index=True)
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
x = new_data.drop('Class',axis=1)
y = new_data['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

# **LOGISTIC REGRESSION**

```python
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred1 = log.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test, y_pred1))
accuracy_score(y_test,y_pred1)
```

//  The Logostic Regression Model gives 94% accuracy.

## **DECISION TREE CLASSIFIER**

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred2 = dt.predict(x_test)
```

```python
print(classification_report(y_test, y_pred2))
accuracy_score(y_test,y_pred2)
```

//  The Decision Tree Classifier Model gives 90% accuracy.

## **RANDOM FOREST CLASSIFIER**

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred3 = rf.predict(x_test)
print(classification_report(y_test, y_pred3))
accuracy_score(y_test,y_pred3)
```

//  The Random Forest Classifier Model gives 94% accuracy.

```python
final_data = pd.DataFrame({'Models':['LR','DT','RF'],"ACC":[accuracy_score(y_test,y_pred1)*100,accuracy_score(y_test,y_pred2)*100,accuracy_score(y_test,y_pred3)*100]})
```
